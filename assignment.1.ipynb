```python
### Assignment Questions

#### Q1: Explain the following with an example:
1. **Artificial Intelligence (AI)**:
   - **Definition**: AI is the simulation of human intelligence processes by machines, especially computer systems. These processes include learning, reasoning, and self-correction.
   - **Example**: A chatbot that can understand and respond to customer queries, such as Apple's Siri or Amazon's Alexa.

2. **Machine Learning (ML)**:
   - **Definition**: ML is a subset of AI that involves the use of algorithms and statistical models to enable a computer to improve its performance on a specific task through experience (data) without being explicitly programmed.
   - **Example**: An email spam filter that learns to classify emails as spam or not based on previous emails labeled by users.

3. **Deep Learning (DL)**:
   - **Definition**: DL is a subset of ML that uses neural networks with many layers (deep neural networks) to analyze various factors of data. DL is especially powerful for tasks such as image and speech recognition.
   - **Example**: A deep learning model used in autonomous vehicles to identify objects like pedestrians, cars, and traffic signs.

#### Q2: What is supervised learning? List some examples of supervised learning.
- **Definition**: Supervised learning is a type of machine learning where the model is trained on labeled data. The input data is paired with the correct output, and the model learns to map the inputs to the outputs.
- **Examples**:
  - Regression: Predicting house prices based on features like size, location, etc.
  - Classification: Email spam detection, where emails are labeled as 'spam' or 'not spam'.
  - Object recognition in images: Identifying objects in a photo with labeled examples of each object.

#### Q3: What is unsupervised learning? List some examples of unsupervised learning.
- **Definition**: Unsupervised learning is a type of machine learning where the model is trained on unlabeled data. The model tries to learn the patterns and the structure from the input data without any explicit outputs provided.
- **Examples**:
  - Clustering: Grouping customers based on purchasing behavior without predefined categories.
  - Dimensionality Reduction: Reducing the number of features in a dataset while retaining important information, such as using Principal Component Analysis (PCA).
  - Anomaly detection: Identifying unusual data points within a dataset.

#### Q4: What is the difference between AI, ML, DL, and DS?
- **AI (Artificial Intelligence)**: The broader concept of machines being able to carry out tasks in a way that we would consider “smart.”
- **ML (Machine Learning)**: A subset of AI that involves the use of algorithms to allow computers to learn from and make decisions based on data.
- **DL (Deep Learning)**: A subset of ML that uses neural networks with many layers to analyze various data representations.
- **DS (Data Science)**: An interdisciplinary field that uses scientific methods, processes, algorithms, and systems to extract knowledge and insights from structured and unstructured data.

#### Q5: What are the main differences between supervised, unsupervised, and semi-supervised learning?
- **Supervised Learning**:
  - Trained on labeled data.
  - Goal is to learn a mapping from inputs to outputs.
  - Examples: Classification, Regression.

- **Unsupervised Learning**:
  - Trained on unlabeled data.
  - Goal is to identify patterns or structures in the data.
  - Examples: Clustering, Dimensionality Reduction.

- **Semi-Supervised Learning**:
  - Trained on a small amount of labeled data and a large amount of unlabeled data.
  - Combines the advantages of both supervised and unsupervised learning.
  - Example: Using a small labeled dataset to guide the learning process on a larger unlabeled dataset.

#### Q6: What is train, test and validation split? Explain the importance of each term.
- **Training Set**:
  - The portion of the data used to train the model.
  - The model learns patterns and relationships within this data.

- **Validation Set**:
  - The portion of the data used to tune the model's hyperparameters.
  - It helps in preventing overfitting and in selecting the best model configuration.

- **Test Set**:
  - The portion of the data used to evaluate the final model's performance.
  - It provides an unbiased assessment of the model's accuracy on new, unseen data.

#### Q7: How can unsupervised learning be used in anomaly detection?
- **Unsupervised learning** can identify patterns and structures in the data without prior labels. In anomaly detection, it can be used to find data points that deviate significantly from the majority of the data.
- **Examples**:
  - **Clustering Algorithms**: Clusters the data and identifies points that do not fit well into any cluster.
  - **Density Estimation**: Identifies low-density regions in the data which correspond to anomalies.

#### Q8: List down some commonly used supervised learning algorithms and unsupervised learning algorithms.
- **Supervised Learning Algorithms**:
  - Linear Regression
  - Logistic Regression
  - Decision Trees
  - Random Forest
  - Support Vector Machines (SVM)
  - Neural Networks

- **Unsupervised Learning Algorithms**:
  - K-Means Clustering
  - Hierarchical Clustering
  - Principal Component Analysis (PCA)
  - Independent Component Analysis (ICA)
  - DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
  - Autoencoders
```
